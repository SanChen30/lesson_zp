```python
def send_message(messages):
    response = client.chat.completions.create(
        model='deepseek-reasoner',
        messages=messages,
        # 告诉大模型可以调用tools
        tools=tools,
        # 让大模型自由选择
        tool_choice='auto'
    )
    return response
```

```python
# openai 定义的 tool 接口标准
tools = [
    # 创建tool JSON声明
    {
        # tool 就是一个函数
        "type": "function",
        "function": {
            "name": "get_closing_price",
            # 自然语言处理，根据自然语言 auto 选择 tool
            "description": "获得指定股票的收盘价",
            # 参数
            "parameters": {
                "type": "object",
                # 接口
                "properties": {
                    "name": {
                        "type": "string",
                        "description": "股票名称"
                    }
                },
                "required": ["name"]
            }
        }
    }
]
```

```python
def get_closing_price(name):
    if name == '青岛啤酒':
        return '67.92'
    else:
        return '未找到该股票'
```

```python
messages = [{"role": "user", "content": "青岛啤酒的收盘价是多少？"}]
response = send_message(messages)
print(response.choices[0].message.content)
```
