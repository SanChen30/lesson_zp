```python
def send_message(messages):
    response = client.chat.completions.create(
        model='deepseek-reasoner',
        messages=messages
    )
    return response

messages = [{'role': 'user', 'content': '青岛啤酒的收盘价是多少？'}]
response = send_message(messages)
print(response.choices[0].message.content)
```
